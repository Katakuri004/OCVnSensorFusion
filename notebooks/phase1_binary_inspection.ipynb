# Phase 1 Binary Inspection

Use this notebook to validate `.imu` and `.vts` assumptions before full rendering.

```python
from pathlib import Path
import struct
import binascii

imu_path = Path('../data/recording2.imu')
vts_path = Path('../data/recording2.vts')

with imu_path.open('rb') as f:
    imu_magic = f.read(8)
    imu_preview = f.read(64)

with vts_path.open('rb') as f:
    vts_magic = f.read(8)
    vts_preview = f.read(24)

print('imu magic:', imu_magic)
print('vts magic:', vts_magic)
print('imu preview:', binascii.hexlify(imu_preview))
print('vts preview:', binascii.hexlify(vts_preview))
```
